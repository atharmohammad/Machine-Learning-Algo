# Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_features=2,n_samples=500,centers=2,random_state=11)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

# Model and Helper Function

```python
def sigmoid(z):
    return 1.0/(1+np.exp(-z))
```

```python
def predict(X,weights):
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions
```

```python
def loss(X,Y,weights):
    Y_ = predict(X,weights)
    cost = np.mean(-Y*np.log(Y_) + (1-Y)*np.log(1-Y_))
    return cost
```

```python
def update(X,Y,weights,learning_rate):
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_-Y)
    
    m = X.shape[0]
    weights = weights - learning_rate*dw/float(m)
    return weights
```

```python
def train(X,Y,learning_rate=0.01,max_itr=5000):
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    weights = np.zeros(X.shape[1])
    losses = []
    for epoch in range(max_itr):
        weights = update(X,Y,weights,learning_rate)
        
        l = loss(X,Y,weights)
        losses.append(l)
            
    
    return weights,losses
```

```python
weights,losses = train(X,Y)
```

```python
plt.title("Loss change")
plt.plot(losses)
plt.show()
```

```python
print(weights)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
x1 = np.linspace(-10,4,10)
x2 = -(weights[1]*x1 + weights[0])/weights[2]
plt.plot(x1,x2)
plt.show()
```

```python
def get_predictions(X_test,weights,label=False):
    if X_test.shape[1] != weights.shape[0]:
        ones = np.ones((X_test.shape[0],1))
        X_test = np.hstack((ones,X_test))
        
        probs = predict(X_test,weights)
        
        if label == True:
            labels = np.zeros(probs.shape,dtype='uint8')
            labels[probs>=.5] = 1
            return labels
        else:
            return probs
        
```

```python
Y_ = get_predictions(X,weights,label=True)
```

```python
acc = np.sum(Y_==Y)/Y.shape[0]
```

```python
print(f"Accuracy of the prediction is  {acc*100}")
```

# Predicting Non-linear dataset

```python
from sklearn.datasets import make_moons
```

```python
X,Y = make_moons(n_samples=500,noise=0.1,random_state=11)
```

```python
plt.title("Data set")
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
weights,losses = train(X,Y)
```

```python
plt.title("Classifying the data set")
plt.scatter(X[:,0],X[:,1],c=Y)
x1 = np.linspace(-1,2,10)
x2 = -(weights[1]*x1 + weights[0])/weights[2]
plt.plot(x1,x2)
plt.show()
```

```python
Y_ = get_predictions(X,weights,label=True)
```

```python
acc = np.sum(Y_==Y)/Y.shape[0]
```

```python
print(f"Accuracy of the prediction is  {acc*100}")
```
