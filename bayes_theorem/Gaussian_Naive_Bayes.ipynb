```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
gnb = GaussianNB()
```

```python
X,Y = make_classification(n_samples=200,n_features=2,n_informative=2,n_redundant=0,random_state=36)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
print(X.shape,Y.shape)
```

```python
gnb.fit(X,Y)
```

```python
ypred = gnb.predict(X)
print(ypred)
```

```python
score = gnb.score(X,Y)
print(score)
```

```python
acc = np.sum(ypred==Y)/Y.shape[0]
print(acc)
```
