```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from pathlib import Path
```

```python
p = Path('./Temp/')
d = p.glob('*')
image_data = []

labels_dict = {
    "Pikachu":0,
    "Bulbasaur":1,
    "Meowth":2
}

label2pokemon = {
    0:"Pikachu",
    1:"Bulbasaur",
    2:"Meowth"
}

labels = []

image_paths = []
```

```python
for folder_dir in d:
    lab = str(folder_dir).split("\\")[-1]
    for img_path in folder_dir.glob('*.jpg'):
        img = image.load_img(img_path,target_size=(40,40))
        img_array = image.img_to_array(img)
        image_data.append(img_array)
        labels.append(labels_dict[lab])
        
```

## Converting data in numpy arrays

```python
from sklearn.utils import shuffle
X = np.array(image_data)
Y = np.array(labels)
X,Y = shuffle(X,Y,random_state=2)

X = X/255.0 #Normalisation , Since values are float so for RGB we convert them in range 0..1

print(X.shape)
print(Y.shape)
```

## Creating test and train data

```python
split = int(0.8*X.shape[0])

Xtrain = X[:split,:]
Ytrain = Y[:split]

Xtest = X[split:,:]
Ytest = Y[split:]

print(Xtrain.shape,Ytrain.shape)
print(Xtest.shape,Ytest.shape)
```

```python
def softmax(a):
    e_a = np.exp(a)
    ans = e_a/np.sum(e_a,axis=1,keepdims=True)
    return ans
```

```python
class NeuralNetwork:
    
    def __init__(self,input_size,layers,output_size):
        np.random.seed(0)
        model = {}
        model['W1'] = np.random.randn(input_size,layers[0])
        model['b1'] = np.zeros((1,layers[0]))
        
        model['W2'] = np.random.randn(layers[0],layers[1])
        model['b2'] = np.zeros((1,layers[1]))
        
        model['W3'] = np.random.randn(layers[1],output_size)
        model['b3'] = np.zeros((1,output_size))
        
        self.model = model
    
    def forward(self,X):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        b1,b2,b3 = self.model['b1'],self.model['b2'],self.model['b3']
        
        Z1 = np.dot(X,W1) + b1
        a1 = np.tanh(Z1)
        
        Z2 = np.dot(a1,W2) + b2
        a2 = np.tanh(Z2)
        
        Z3 = np.dot(a2,W3) + b3
        Y_ = softmax(Z3)
        self.acitvation_output = (a1,a2,Y_)
        return Y_
    
    def backward(self,X,Y,lr=0.001):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        b1,b2,b3 = self.model['b1'],self.model['b2'],self.model['b3']
        
        a1,a2,Y_ = self.acitvation_output
        m = X.shape[0]
        
        delta3 = Y_ - Y
        
        dw3 = np.dot(a2.T,delta3)
        db3 = np.sum(delta3,axis=0)
        
        delta2 = (1-np.square(a2))*np.dot(delta3,W3.T)
        dw2 = np.dot(a1.T,delta2)
        db2 = np.sum(delta2,axis=0)
        
        delta1 = (1-np.square(a1))*np.dot(delta2,W2.T)
        dw1 = np.dot(X.T,delta1)
        db1 = np.sum(delta1,axis=0)
        
        self.model['W1'] -= lr*dw1
        self.model['b1'] -= lr*db1
        
        self.model['W2'] -= lr*dw2
        self.model['b2'] -= lr*db2
        
        self.model['W3'] -= lr*dw3
        self.model['b3'] -= lr*db3
        
    def predict(self,X):
        Y_out = self.forward(X)
        return np.argmax(Y_out,axis=1)
    
    def summary(self):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        b1,b2,b3 = self.model['b1'],self.model['b2'],self.model['b3']
        a1,a2,Y_ = self.acitvation_output
        
        print('W1 : ',W1.shape)
        print('A1 : ',a1.shape)
        
        print('W2 : ',W2.shape)
        print('A2 : ',a2.shape)
        
        print('W3 : ',W3.shape)
        print('YPred : ',Y_.shape)
```

```python
def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p))
    return l    

def one_hot_vector(y,depth):
    m = y.shape[0]
    y_oht = np.zeros((m, depth))
    y_oht[np.arange(m), y] = 1
    return y_oht
```

```python
def train(X,Y,model,epochs,lr=0.0001,logs=True):
    training_loss = []
    
    classes = 3
    Y_OHT = one_hot_vector(Y,classes)
    
    for ix in range(epochs):
        Y_ = model.forward(X)
        l = loss(Y_OHT,Y_)
        training_loss.append(l)
        model.backward(X,Y_OHT,lr)
        if logs and ix%50 == 0:
            print(f"Loss in epoch {ix} is : {l}")
        
    return training_loss
```

```python
model = NeuralNetwork(input_size=4800,layers=[100,50],output_size=3) 
```

```python
Xtrain = Xtrain.reshape((Xtrain.shape[0],-1))
Xtest = Xtest.reshape((Xtest.shape[0],-1))
```

```python
Xtrain.shape
```

```python
losses = train(Xtrain,Ytrain,model,500)
```

```python
plt.title("losses descent in training")
plt.plot(losses)
plt.show()
```

```python
test_output = model.predict(Xtest)
train_output = model.predict(Xtrain)
```

```python
test_accuracy = np.sum(test_output==Ytest)/Ytest.shape[0]
train_accuracy = np.sum(train_output==Ytrain)/Ytrain.shape[0]
print(f"training dataset accuracy is : {train_accuracy} and testing dataset accuracy is : {test_accuracy}")
```

## Hence we can see that the model is overfitting as training accuracy is much higher than the testing accuracy
