# Linear-Regression And its Visualization

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
X_train = pd.read_csv("./Data/Training_Data/Linear_X_Train.csv")
Y_train = pd.read_csv("./Data/Training_Data/Linear_Y_Train.csv")
```

```python
X = X_train.values
Y = Y_train.values
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.show()
```

### Normalising the data

```python
u = X.mean()
std = X.std()
X = (X-u)/std
plt.scatter(X,Y)
plt.show()
```

```python
def hypothesis(theta,x):
    return theta[0] + theta[1]*x
```

```python
def gradient(X,Y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    for i in range(m):
        Ypred = hypothesis(theta,X[i])
        grad[0] += (Ypred - Y[i])
        grad[1] += (Ypred - Y[i])*X[i]
    
    return grad/m
```

```python
def error(X,Y,theta):
    m = X.shape[0]
    e = 0
    for i in range(m):
        Ypred = hypothesis(theta,X[i])
        e += (Ypred - Y[i])**2
        
    return e/m
```

```python
def gradient_descent(X,Y,max_steps=100,lr=0.1):
    theta = np.zeros((2,)) # taking the value of m,c  as 0,0 randomly for the initialisation
    error_list = []
    theta_list = []
    m = X.shape[0]
    for i in range(max_steps):
        grad = gradient(X,Y,theta)
        err = error(X,Y,theta)
        error_list.append(err)
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
        theta_list.append((theta[0],theta[1]))
        
    return theta,error_list,theta_list
```

```python
start = time.time()
theta,error_list,theta_list = gradient_descent(X,Y)
end = time.time()
print("Time taken : " + str(end - start))
theta_list = np.array(theta_list)
error_list = np.array(error_list)
```

```python
print(theta) # best value of c = theta[0]  and m = theta[1]
```

```python
error_list[0:-1:5] # errors are getting minimized and at the minima it becomes almost constant as it is the minimum error we can reach
```

```python
plt.title("error decreasing as we descet")
plt.plot(error_list) 
plt.show()
```

### Plotting the Best Predicted Line

```python
Ypred = hypothesis(theta,X) #finding every ypred using xtrain by y = theta[0] + theta[1]*X for all values of X
plt.title("Best Predicted Line")
plt.scatter(X,Y)
plt.plot(X,Ypred,color="orange",label='Prediction') #Now Plotting every Ypred with X to get best line
plt.legend()
plt.show()
```

## R2 or R - Squared Score

```python
def r2_score(Ypred,Y):
    num = np.sum((Y-Ypred)**2)
    den = np.sum((Y-Y.mean())**2)
    score = 1 - num/den
    return score*100
```

```python
Ypred = hypothesis(theta,X) # finding Y prediction using the best m and c which gives the minimum error
print(r2_score(Ypred,Y))
```

# Linear Regression with Multiple Features

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
X = boston.data
Y = boston.target
```

```python
print(X.shape)
```

```python
df = pd.DataFrame(X)
df.columns = boston.feature_names
df.head()
```

```python
# Normalizing the data
u = np.mean(X,axis=0)
std = np.std(X,axis=0)
X = (X - u)/std
print(X)
```

```python
ones = np.ones((X.shape[0],1)) # adding the dummy column for theta[0]
X = np.hstack((ones,X))

def hypothesis(x,theta):
    m = x.shape[0] # length of feature
    ypred = 0.0
    for i in range(m):
        ypred += theta[i]*x[i]
        
    return ypred

def error(X,Y,theta):
    e = 0.0
    m = X.shape[0]
    
    for i in range(m):
        Ypred = hypothesis(X[i],theta)
        e += (Ypred - Y[i])**2
    
    return e/m


def gradient(X,Y,theta):
    m,n = X.shape
    
    grad = np.zeros((n,)) 
    
    for j in range(n):
        for i in range(m):
            Ypred = hypothesis(X[i],theta)
            grad[j] += (Ypred - Y[i])*X[i][j]
    
    return grad/m


def gradient_descent(X,Y,max_steps=300,lr=0.1):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    for i in range(max_steps):
        e = error(X,Y,theta)
        error_list.append(e)
        
        grad = gradient(X,Y,theta)
        
        for j in range(n):
            theta[j] = theta[j] - lr*grad[j]
    
    return theta,error_list
    
```

```python
start = time.time()
theta,error_list = gradient_descent(X[: , :],Y)
end = time.time()
print("time taken : " + str(end - start))
```

```python
plt.style.use("seaborn")
plt.title("Change in error")
plt.plot(error_list)
plt.show()
```

```python
Ypred = []
for i in range (X.shape[0]):
    pred = hypothesis(X[i],theta)
    Ypred.append(pred)

Ypred = np.array(Ypred)
```

```python
# R-squared score
def r2_score(Ypred,Y):
    num = np.sum((Y-Ypred)**2)
    denom = np.sum((Y.mean()-Y)**2)
    score = 1 - num/denom
    return score*100

r2_score(Ypred,Y)
```

## Efficient Way by Vectorization to increase speed of the funciton

```python
def hypothesisV(X,theta):
    return np.dot(X,theta)

def errorV(X,Y,theta):
    Ypred = hypothesisV(X,theta)
    e = 0.0
    m = X.shape[0]
    e = np.sum((Y-Ypred)**2)
    return e/m

def gradientV(X,Y,theta):
    m = X.shape[0]
    Ypred = hypothesisV(X,theta)
    grad = np.dot(X.T,(Ypred - Y))
    return grad/m

def gradient_descentV(X,Y,lr=0.1,max_evochs=300):
    error_list = []
    m,n = X.shape
    theta = np.zeros((n,))
    for i in range(max_evochs):
        e = errorV(X,Y,theta)
        error_list.append(e)
        
        grad = gradientV(X,Y,theta)
        theta = theta - lr*grad
    
    return theta,error_list
```

```python
start = time.time()
theta,error_list = gradient_descentV(X,Y) 
end = time.time()
print("Time after using vectorization : " + str(end - start))
```

```python
plt.title("Change in error")
plt.plot(error_list)
plt.show()
```
